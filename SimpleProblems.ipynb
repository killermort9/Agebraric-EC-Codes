```python
def greatest_k_for_failTol(failTol,n,q,P):
    for k in range(1,n):
        tau = (n-k)//2
        PfailNew = sum([binomial(n, j)*P**j*(1-P)**(n-j) for j in range(tau+1,n+1)])
        if PfailNew >= failTol:
            print("Bounded: ",tau)
            print("k =", k-1)
            print("rate =",(k-1)/n)
            print(Pfail)
            break
        Pfail = PfailNew
greatest_k_for_failTol(0.02,n,q,P)
greatest_k_for_failTol(0.001,n,q,P)
greatest_k_for_failTol(10**(-10),n,q,P)
```

```python
def Ts(j,l,w,q,n):
    S = 0
    if ((l+j >= w and l+w >= j) and w+j >= l):
        for z in range(0,w+1):
            S += binomial(w,z)*binomial(w-z,z+j-l)*(q-2)**(l+w-j-2*z)*binomial(n-w,n-j-z)*(q-1)**(j+z-w)
    return S

def T(j,w,t,q,n):
    S = 0
    for l in range(0,t+1):
        S += Ts(j,l,w,q,n)
    return S
    
def A(w,k,n,q):
    S = 0
    if (n-k+1 <= w  and w <= n):
        for j in range(0,w-(n-k+1)+1):
            S += (-1)**j*binomial(w,j)*(q**(w-(n-k)-j)-1)
        S *= binomial(n,w)
    elif (w == 0):
        return 1
    return S
    
def Perr(p,q,n,k):
    t = (n-k)//2
    S = 0
    for j in range(t+1,n+1):
        s = 0
        for w in range(1,n):
            s += A(w,k,n,q)*T(j,w,t,q,n)
        S += (p/(q-1))**j*(1-p)**(n-j)*s
    return S
```
